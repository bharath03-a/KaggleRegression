### Feature Engineering

```python
# split engine columns
def clean_engine(df, column_name):
    df['hp'] = None
    df['vol'] = None
    df['cylinders'] = None

    for index, row in df.iterrows():
        input_string = row[column_name]
        
        hp_match = re.search(r'(\d+(\.\d+)?)HP', input_string)
        liter_match = re.search(r'(\d+(\.\d+)?)L', input_string)
        cylinder_match = re.search(r'(\d+)\s+Cylinder', input_string)
        fuel_match = re.search(r'(\w+)\s+Fuel', input_string)

        if hp_match:
            df.at[index, 'hp'] = hp_match.group(1)

        if liter_match:
            df.at[index, 'vol'] = liter_match.group(1)

        if cylinder_match:
            df.at[index, 'cylinders'] = cylinder_match.group(1)
            
        if pd.isna(row['fuel_type']) and fuel_match:
            df.at[index, 'fuel_type'] = fuel_match.group(1)

    df = df.drop(columns=[column_name])
    
    return df
```

```python
train_data = clean_engine(train_data, 'engine')
test_data = clean_engine(test_data, 'engine')
```

```python
train_data['price'] = np.log10(train_data['price'])
train_data['milage'] = np.log10(train_data['milage'])
```

```python
# split transmission columns
def clean_transmission(val):
    speed_match = re.search(r'(\d+)-Speed', val)
    speeds = int(speed_match.group(1)) if speed_match else None
    
    if 'A/T' in val or 'automatic' in val.lower():
        transmission_type = 'Automatic' 
    elif 'M/T' in val or 'manual' in val.lower():
        transmission_type = 'Manual'
    else:
        transmission_type = 'Other'
    
    return transmission_type, speeds
```

```python
train_data[['transmission_type', 'speeds']] = train_data['transmission'].apply(lambda x: pd.Series(clean_transmission(x)))
test_data[['transmission_type', 'speeds']] = test_data['transmission'].apply(lambda x: pd.Series(clean_transmission(x)))
```

## Model Training

```python
x = train_data.drop('price', axis=1)
y = train_data['price']

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=random_seed)
```

```python
class regressionModel:
    def __init__(self, model, X_train, X_test, y_train, y_test):
        self.X_train = X_train
        self.X_test = X_test
        self.y_train = y_train
        self.y_test = y_test
        self.model = model
        
    def model_training(self):
        self.model.fit(self.X_train, self.y_train)
        
        y_train_pred = self.model.predict(self.X_train)
        y_test_pred = self.model.predict(self.X_test)
        
        return y_train_pred, y_test_pred
    
    def adjusted_r2(self, r2, n, k):
        return 1 - (1 - r2) * ((n - 1) / (n - k - 1))
    
    def model_evaluation(self, train_pred, test_pred):
        train_mse = mean_squared_error(self.y_train, train_pred)
        test_mse = mean_squared_error(self.y_test, test_pred)
        train_rmse = np.sqrt(train_mse)
        test_rmse = np.sqrt(test_mse)
        train_mae = mean_absolute_error(self.y_train, train_pred)
        test_mae = mean_absolute_error(self.y_test, test_pred)
        test_r2 = r2_score(self.y_test, test_pred)
        train_r2 = r2_score(self.y_train, train_pred)

        n_test = self.X_test.shape[0]
        k = self.X_test.shape[1]
        adj_test_r2 = self.adjusted_r2(test_r2, n_test, k)

        n_train = self.X_train.shape[0]
        adj_train_r2 = self.adjusted_r2(train_r2, n_train, k)

        print(f"Model : {self.model}")
        print(f"Train MSE: {train_mse}")
        print(f"Test MSE: {test_mse}")
        print(f"Train RMSE: {train_rmse}")
        print(f"Test RMSE: {test_rmse}")
        print(f"Train MAE: {train_mae}")
        print(f"Test MAE: {test_mae}")
        print(f"R^2 Score (Test Data): {test_r2}")
        print(f"R^2 Score (Train Data): {train_r2}")
        print(f"Adjusted R^2 (Test Data): {adj_test_r2}")
        print(f"Adjusted R^2 (Train Data): {adj_train_r2}")
```

```python
# Linear Regression
linear_model = LinearRegression()
linear_regression = regressionModel(linear_model, X_train, X_test, y_train, y_test)
```

```python
%%time
# Linear Regression Training
linear_train_pred, linear_test_pred = linear_regression.model_training()
```

```python
%%time
# Evaluation metrics
linear_regression.model_evaluation(linear_train_pred, linear_test_pred)
```

```python
# Ridge Regression
ridge_model = Ridge()
ridge_regression = regressionModel(ridge_model, X_train, X_test, y_train, y_test)
```

```python
%%time
# Ridge Regression Training
ridge_train_pred, ridge_test_pred = ridge_regression.model_training()
```

```python
%%time
# Evaluation metrics
ridge_regression.model_evaluation(ridge_train_pred, ridge_test_pred)
```

```python
# Lasso Regression
lasso_model = Lasso()
lasso_regression = regressionModel(lasso_model, X_train, X_test, y_train, y_test)
```

```python
%%time
# Lasso Regression Training
lasso_train_pred, lasso_test_pred = lasso_regression.model_training()
```

```python
%%time
# Evaluation metrics
lasso_regression.model_evaluation(lasso_train_pred, lasso_test_pred)
```

```python
# Decision Tree regressor
dt_model = DecisionTreeRegressor()
dt_regression = regressionModel(dt_model, X_train, X_test, y_train, y_test)
```

```python
%%time
# Decision Tree Regression Training
dt_train_pred, dt_test_pred = dt_regression.model_training()
```

```python
%%time
# Evaluation metrics
dt_regression.model_evaluation(dt_train_pred, dt_test_pred)
```

```python
# XGBoost regressor
xgb_model = XGBRegressor()
xgb_regression = regressionModel(xgb_model, X_train, X_test, y_train, y_test)
```

```python
%%time
# XGBoost Regression Training
xgb_train_pred, xgb_test_pred = xgb_regression.model_training()
```

```python
%%time
# Evaluation metrics
xgb_regression.model_evaluation(xgb_train_pred, xgb_test_pred)
```

```python
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.6, 0.8, 1.0],
    'colsample_bytree': [0.6, 0.8, 1.0],
    'gamma': [0, 0.1, 0.2]
}
```

```python
grid_search = GridSearchCV(
    estimator=xgb_model,
    param_grid=param_grid,
    cv=5,
    scoring='neg_mean_squared_error',
    n_jobs=-1,
    verbose=1
)
```

```python
grid_search.fit(X_train, y_train)
```

```python
print("Best Parameters:")
print(grid_search.best_params_)
```

```python
print(f"Best Cross-Validated Negative MSE: {grid_search.best_score_:.4f}")
```

```python
y_pred = grid_search.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Test Set Mean Squared Error: {mse:.4f}")
print(f"Test Set R^2 Score: {r2:.4f}")
```

```python
best_model = grid_search.best_estimator_
```
